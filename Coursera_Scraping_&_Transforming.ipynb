```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

--------------------
# Web Scraping (EXTRACT)

```python
link = "https://www.coursera.org/courses"
response = requests.get(link)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Lists to store the data
course_title = []
course_organization = []
course_Certificate_type = []
course_rating = []
course_reviews = []
coursera_plus = []
difficulty = []
skills_gained = []
course_duration = []

courses_df = pd.DataFrame({
    "Title": course_title,
    "Organization": course_organization,
    "Certificate Type": course_Certificate_type,
    "Rating": course_rating,
    "Reviews": course_reviews,
    "Coursera Plus": coursera_plus,
    "difficulty Level": difficulty,
    "Skills": skills_gained,
    "Duration": course_duration
    })
```

```python
# Loop through all pages
for page in range(1, 85):  # Adjust range if necessary
    print(f"Scraping page {page}...")
    url = f"{link}{page}"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.5735.199 Safari/537.36"
    }
    response = requests.get(url, headers=headers)
    
    if response.status_code != 200:
        print(f"Failed to retrieve page {page}")
        continue

    # Extract course cards
    course_cards = soup.find_all('div', class_='cds-CommonCard-clickArea')
    print(f"No of courses per page: {len(course_cards)}...")
    for course in course_cards:

        # Extract title
        title = course.find('h3', class_='cds-CommonCard-title css-6ecy9b').text.strip() if course.find('h3', class_='cds-CommonCard-title css-6ecy9b') else None
        course_title.append(title  or "")

        # Extract organization
        organization = course.find('p', class_='cds-ProductCard-partnerNames css-vac8rf').text.strip() if course.find('p', class_='cds-ProductCard-partnerNames css-vac8rf') else None
        course_organization.append(organization or "")

        # Extract certificate type
        cert_type = course.find('div', class_='cds-CommonCard-metadata').text.strip() if course.find('div', class_='cds-CommonCard-metadata') else None
        course_Certificate_type.append(cert_type  or "")

        # Extract rating
        rating = course.find('span', class_='css-6ecy9b').text.strip() if course.find('span', class_='css-6ecy9b') else None
        course_rating.append(rating  or "")
        
        # Extract reviews
        review = course.find('div', class_='css-vac8rf').text.strip() if course.find('div', class_='css-vac8rf') else None
        course_reviews.append(review or "")

        # Extract coursera_plus
        Coursera_plus_inc = course.find('div', class_='css-1nvn7oo').text.strip() if course.find('div', class_='css-1nvn7oo') else None
        coursera_plus.append(True or "")

        # Extract difficulty
        difficulty_level = course.find('div', class_='cds-CommonCard-metadata').text.strip() if course.find('div', class_='cds-CommonCard-metadata') else None
        difficulty.append(difficulty_level or "")

        # Extract gained skills
        skills = course.find('div', class_='cds-CommonCard-bodyContent').text.strip() if course.find('div', class_='cds-CommonCard-bodyContent') else None
        skills_gained.append(skills  or "")

        # Extract course duration
        duration = course.find('div', class_='cds-CommonCard-metadata').text.strip() if course.find('div', class_='cds-CommonCard-metadata') else None
        course_duration.append(duration  or "")

# Delay to avoid overwhelming the server
    time.sleep(2)

# Deqbugging to see if the data was scraped

print(f"Titles: {len(course_title)}")
print(f"Organizations: {len(course_organization)}")
print(f"Certificate Types: {len(course_Certificate_type)}")
print(f"Ratings: {len(course_rating)}")
print(f"Reviews: {len(course_reviews)}")
print(f"Coursera Plus: {len(coursera_plus)}")
print(f"difficulty: {len(difficulty)}")
print(f"Skills: {len(skills_gained)}")
print(f"Duration: {len(course_duration)}")


# Create a DataFrame
courses_df = pd.DataFrame({
    "Title": course_title,
    "Organization": course_organization,
    "Certificate Type": course_Certificate_type,
    "Rating": course_rating,
    "Reviews": course_reviews,
    "Coursera Plus": coursera_plus,
    "difficulty": difficulty,
    "Skills": skills_gained,
    "Duration": course_duration
    
})

# Save to CSV
courses_df.to_csv('data/coursera_courses.csv', index=False)
print("Scraping completed. Data saved to 'coursera_courses.csv'.")
```

```python
courses_df_clean["Certificate Type"] = courses_df["Certificate Type"].str.split('Â·').str[0].str.strip()
```

```python
courses_df_clean["difficulty"] = courses_df["difficulty"].str.split('Â·').str[1].str.strip()
```

```python
courses_df_clean["Duration"] = courses_df["Duration"].str.split('Â·').str[2].str.strip()
```

```python
courses_df_clean["Skills"] = courses_df["Skills"].str.split("Skills you'll gain:").str[1].str.strip()
```

```python
courses_df_clean["Skills"] = courses_df["Skills"].str.split(",")
```

```python
def convert_reviews_int(reviews):
    #removing text
    reviews = reviews.replace(" reviews","")
    if 'K' in reviews:
        return int(float(reviews.replace('K',""))* 1000)
    return int(reviews)
```

```python
courses_df_clean["Reviews"] = courses_df_clean["Reviews"].apply(convert_reviews_int)
```
