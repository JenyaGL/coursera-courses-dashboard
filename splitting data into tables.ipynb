# creating tables
* setting up unique keys and preparing to load to databases

```python
data=pd.read_csv('data/clean_coursera_courses.csv')
data.columns=data.columns.str.lower()
```

```python
data['unique_id'] = range(1, len(data) + 1)
```

```python
data
```

```python
courses = data[['unique_id', 'title', 'organization']].drop_duplicates()
courses
```

```python
# Renaming columns
courses= courses.rename({'unique_id':"course_id"}, axis='columns')
courses
```

```python
courses_dim = courses
courses_dim
```

```python
courses_dim.to_csv('data/courses_dim.csv')
```

```python
course_details = data[['unique_id', 'certificate type', 'rating', 'reviews', 'coursera plus', 'difficulty']].drop_duplicates()
course_details= course_details.rename({'unique_id':"course_id"}, axis='columns')
```

```python
course_details
```

```python
course_details_dim = course_details
course_details_dim
```

```python
course_details_dim.to_csv('data/course_details_dim.csv')
```

```python
# adjust split(,) if needed
skills_series = data['skills'].dropna().apply(lambda x: x.split(','))
skills = pd.Series([skill.strip() for sublist in skills_series for skill in sublist]).drop_duplicates().reset_index(drop=True)
skills = skills.to_frame(name='skill_name')
skills['skill_id'] = skills.index + 1  # Assign a unique skill_id
skills_list = skills
```

```python
skills['skill_name'] = skills['skill_name'].str.strip("]").str.replace("'", "")
skills['skill_name'] = skills['skill_name'].str.strip("[").str.replace("'", "")
skills['skill_name'] = skills['skill_name'].str.strip(" ")
```

```python
skills_list_dim = skills_list
skills_list_dim
```

```python
skills_list_dim.to_csv('data/skills_list_dim.csv')
```
