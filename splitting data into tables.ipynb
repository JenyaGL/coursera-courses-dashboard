# Creating tables
* setting up unique keys for every course
* seperating the coulmn into different dimentions:
  
   - courses - consisti of unique ID, name and organization
   - course_details - consist of course_id, certification, rating, reviews, coursera plus, included in degree points and difficulty
   - skills list - that will consist of skill name and skill id
   - course to skill - consist of id, skill idand course id
    
* preparing to load to databases

```python
data=pd.read_csv('data/clean_coursera_courses.csv')
data.columns=data.columns.str.lower()
```

```python
data['unique_id'] = range(1, len(data) + 1)
```

```python
data
```

```python
courses = data[['unique_id', 'title', 'organization']].drop_duplicates()
courses
```

```python
# Renaming columns
courses= courses.rename({'unique_id':"course_id"}, axis='columns')
courses
```

```python
courses_dim = courses
courses_dim
```

```python
courses_dim.to_csv('data/courses_dim.csv', index=False)
```

```python
course_details = data[['unique_id', 'certificate type', 'rating', 'reviews', 'coursera plus', 'towards degree', 'difficulty']].drop_duplicates()
course_details= course_details.rename({'unique_id':"course_id",'towards degree':'degree points'}, axis='columns')
```

```python
course_details
```

```python
course_details_fact = course_details
course_details_fact
```

```python
course_details_fact.to_csv('data/course_details_fact.csv',index=False)
```

```python
# adjust split(,) if needed
skills_series = data['skills'].dropna().apply(lambda x: x.split(','))
skills = pd.Series([skill.strip() for sublist in skills_series for skill in sublist]).drop_duplicates().reset_index(drop=True)
skills = skills.to_frame(name='skill_name')
skills['skill_id'] = skills.index + 1  # Assign a unique skill_id
skills_list = skills
```

```python
skills['skill_name'] = skills['skill_name'].str.strip("]").str.replace("'", "")
skills['skill_name'] = skills['skill_name'].str.strip("[").str.replace("'", "")
skills['skill_name'] = skills['skill_name'].str.strip(" ")
```

```python
skills_list_dim = skills_list
skills_list_dim
```

```python
skills_list_dim.to_csv('data/skills_list_dim.csv',index=False)
```

```python
data
```

```python
data['skills'] = data['skills'].str.strip("]").str.replace("'", "")
data['skills'] = data['skills'].str.strip("[").str.replace("'", "")
data['skills'] = data['skills'].str.strip(" ")
```

```python
data['skills']
```

```python
exploded_skills = data.explode('skills')  # Create one row per skill
exploded_skills['skills'] = exploded_skills['skills'].str.strip()  # Remove extra spaces
```

```python
# Merge with skills_list_dim to get skill_id
course_to_skill_dim = exploded_skills.merge(skills_list_dim, left_on='skills', right_on='skill_name')
```

```python
# Merge with courses_dim to get course_id
course_to_skill_dim = course_to_skill_dim.merge(courses_dim, on='title')
```

```python
course_to_skill_dim
```

```python
# Select relevant columns
course_to_skill_dim = course_to_skill_dim[['course_id', 'skill_id']]
```

```python
# Optionally, add a unique identifier for each row
course_to_skill_dim['id'] = range(1, len(course_to_skill_dim) + 1)

# Reorder columns
course_to_skill_dim = course_to_skill_dim[['id', 'course_id', 'skill_id']]

# Display the final dimension table
course_to_skill_dim
```

```python
# Save the dimension table to a CSV file
course_to_skill_dim.to_csv('data/course_to_skill_dim.csv',index=False)
```
